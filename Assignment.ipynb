```python
# Import libraries
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from tqdm import tqdm
from joblib import Parallel, delayed
```

```python
# Get the HTML content of the page
link = "https://ic2s2-2023.org/program"
r = requests.get(link)
soup = BeautifulSoup(r.content)
```

```python
table = soup.find("table",{"class":"tutorials"})
table_rows = table.find_all("tr")

# Here we find the keynote talks and store them in a list wow
keynotes = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [i.text for i in td]
    if row:
        if "Keynote" in row[1]:
            keynotes.append(row[1])

print(keynotes)

# cleaning it up so we only have the names of the keynotes
keynotes = [re.sub("Keynote - ", "", i) for i in keynotes]

print(keynotes)
```

```python
# we saw a pattern and get these so we get some chairs
pattern = r'Chair:\s+([A-Za-z\s]+?)(?=[,.\n])'

# Find all occurrences of the pattern
matches = re.findall(pattern, soup.get_text())

# Processing the matches to trim any trailing whitespace
chairs = [match.strip() for match in matches]

# Print the list or process as needed
print(chairs)
```

```python
names = []

# Find all <i> tags
italic_tags = soup.find_all('i')

# Extract names from each <i> tag because we saw it was in italics (but here we also have the chairs)
for tag in italic_tags:
    # Split the text by commas and strip whitespace
    names_in_tag = [name.strip() for name in tag.get_text().split(',')]
    names.extend(names_in_tag)

# Print the list or process as needed
print(names)

# Remove element in list if it starts with "Chair:" because otherwise we get a lot of duplicates
authors_posters = [name for name in names if not name.startswith('Chair:')]
print(authors_posters)
```

```python
print(len(keynotes))
print(len(chairs))  
print(len(authors_posters)) 

all = keynotes + chairs + authors_posters

print(len(all))
#find unique elements
all = set(all)
print(len(all))

df = pd.DataFrame(all, columns=['Name'])
df.to_csv('names.txt', index=False)
```

# Part 3 Gathering Research Articles using the OpenAlex API

```python
# Load the dataframe created in week 2 
df_authors = pd.read_csv('authors.csv')
display(df_authors)
```

```python
# we get the id's so we can filter using these

concepts = ['Sociology', 'Psychology', 'Economics', 'Political science', 'Mathematics', 'Physics', 'Computer science']
url = "https://api.openalex.org/concepts"
concept_id = []
for concept in concepts:
    response = requests.get(url, params={'search': concept, 'filter': 'level:0'})
    data = response.json()
    concept_id.append(data['results'][0]['id'])
```

```python
df_authors = df_authors[(df_authors['works_count'] > 5) & (df_authors['works_count'] < 5000)]

# Request function for parallel processing
def request_works(author_ids: list[str]) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Request works from the OpenAlex API for a list of authors.
    
    Args:
    author_names: list of author names (str)

    Returns:
    df_papers: DataFrame with the works of the authors
    df_abstracts: DataFrame with the abstracts of the works
    """
    # Initialize empty DataFrames for storing the data
    df_papers = pd.DataFrame()
    df_abstracts = pd.DataFrame()

    # For 20 authors at a time, make a request to the OpenAlex API to fetch their works
    author_id = "|".join(author_ids)
    url = "https://api.openalex.org/works?"
    params = {'per-page': 200, 
        'cursor': '*',
        'filter': f'authorships.author.id:{author_id}' + ',' + 
        'cited_by_count:>10' + ',' + 
        'authors_count:<10' + ',' +
        f'concept.id:{concept_ids[0]}|{concept_ids[1]}|{concept_ids[2]}|{concept_ids[3]}' + ',' +
        f'concept.id:{concept_ids[4]}|{concept_ids[5]}|{concept_ids[6]}' 
        }
    response = requests.get(url, params=params)
    data = response.json()

    # For each work, extract the relevant information and add it to the DataFrames, while going through the pages
    while data['meta']['next_cursor'] != None:
        if data['meta']['next_cursor'] == 0:
            break
        else:
            params["cursor"] = data['meta']["next_cursor"]
            for work in data['results']:
                
                df_papers_current = pd.DataFrame({
                    'id': [work['id']],
                    'publication_year': [work['publication_year']],
                    'cited_by_count': [work['cited_by_count']],
                    'author_ids': [[author['author']['id'] for author in work['authorships']]]
                })
                df_papers = pd.concat([df_papers, df_papers_current], ignore_index=True)

                df_abstracts_current = pd.DataFrame({
                    'id': [work['id']],
                    'title': [work['title']],
                    'abstract_inverted_index': [work['abstract_inverted_index']]
                })
                df_abstracts = pd.concat([df_abstracts, df_abstracts_current], ignore_index=True)

            response = requests.get(url, params=params)
            data = response.json()
    
    return df_papers, df_abstracts

# Parallel processing
results = Parallel(n_jobs=4)(delayed(request_works)(df_authors['id'][i:i+20]) for i in tqdm(range(0, len(df_authors['id']), 20), desc='Requesting data: '))
  
```

```python
df_papers = pd.concat([result[0] for result in results], ignore_index=True)
df_abstracts = pd.concat([result[1] for result in results], ignore_index=True)

print(df_papers.shape)
print(df_abstracts.shape)

# Save the DataFrames to CSV files
df_papers.to_csv('papers.csv', index=False)
df_abstracts.to_csv('abstracts.csv', index=False)
```

```python
df_authors = pd.read_csv('author_ids.csv')
df_authors = df_authors.rename(columns={'id': 'author_id'})

# For every author id in the author ids column of df_papers, give the author id a row in the dataframe
df_expanded_papers = df_papers.explode('author_ids').rename(columns={'author_ids': 'author_id'})

# Unique author ids
unique_author_ids = df_expanded_papers['author_id'].unique()
print('Unique autor IDs: ',len(unique_author_ids))

# Load
df_coauthors = pd.read_csv('co-authors.csv')
df_coauthors = df_coauthors.rename(columns={'id': 'author_id'})

# Append df_authors to df_coauthors
df_authors_combined = pd.concat([df_coauthors, df_authors])

# Remove duplicate author ids
df_authors_combined = df_authors_combined.drop_duplicates(subset='author_id')

# Merge the two dataframes
df_merged = pd.merge(df_expanded_papers, df_authors_combined, on='author_id')
```

# Part 4: The Network of Computational Social Scientists

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import os
from collections import defaultdict
import ast
import json
```

Getting the dict

```python
# Load the papers.csv file
df_papers = pd.read_csv('papers.csv')
df_papers['author_ids'] = df_papers['author_ids'].apply(ast.literal_eval)

# Construct the adjacency list of the graph so we dont have to initialize when a new pair is found. 
adj_list = defaultdict(int)

for i in range(len(df_papers)):
    authors = df_papers['author_ids'][i]
    for j in range(len(authors)):
        for k in range(j+1, len(authors)):
            adj_list[(authors[j], authors[k])] += 1
```

Creating the graph

```python
# Construct the graph
G = nx.Graph()
weighted_edge_list = [(u, v, d) for (u, v), d in adj_list.items()]
G.add_weighted_edges_from(weighted_edge_list)
```

Check if we still authors is loaded

```python
# Add information to the graph nodes
for i in range(len(df_merged)):
    author_id = df_merged['author_id'][i]
    # Add the information to the graph if the node exists
    if author_id in G.nodes:
        G.nodes[author_id]['display_name'] = df_merged['display_name'][i]
        G.nodes[author_id]['country_code'] = df_merged['country_code'][i]
        G.nodes[author_id]['cited_by_count'] = int(df_merged['cited_by_count'][i])
        G.nodes[author_id]['publication_year'] = int(df_merged['publication_year'][i])

# Save the graph as json
graph_data = nx.node_link_data(G)
json.dump(graph_data, open('graph.json', 'w'))
```

1. Network Metrics:

```python
# Total number of nodes (authors)
num_nodes = G.number_of_nodes()

# Total number of links (collaborations)
num_edges = G.number_of_edges()

# Network's density
density = nx.density(G)

# Check if the network is fully connected
is_connected = nx.is_connected(G)

# Number of connected components
num_connected_components = nx.number_connected_components(G)

# Number of isolated nodes
num_isolated_nodes = nx.number_of_isolates(G)

print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
print(f"Density: {density}")
print(f"Is the network fully connected? {'Yes' if is_connected else 'No'}")
print(f"Number of connected components: {num_connected_components}")
print(f"Number of isolated nodes: {num_isolated_nodes}")
```

2. Degree Analysis:

```python
# Get the degrees of the nodes
degrees = [deg for node, deg in G.degree()]

# Compute the statistics for degree
avg_degree = np.mean(degrees)
median_degree = np.median(degrees)
degree_uniq, degree_count = np.unique(degrees, return_counts=True)
mode_degree = degree_uniq[np.argmax(degree_count)]
min_degree = np.min(degrees)
max_degree = np.max(degrees)

# Get the weighted degrees (node strengths) of the nodes
strengths = [strength for node, strength in G.degree(weight='weight')]

# Compute the statistics for node strength
avg_strength = np.mean(strengths)
median_strength = np.median(strengths)
strength_uniq, strength_count = np.unique(strengths, return_counts=True)
mode_strength = strength_uniq[np.argmax(strength_count)]
min_strength = np.min(strengths)
max_strength = np.max(strengths)

print(f"Degree: Avg={avg_degree}, Median={median_degree}, Mode={mode_degree}, Min={min_degree}, Max={max_degree}")
print(f"Node strength: Avg={avg_strength}, Median={median_strength}, Mode={mode_strength}, Min={min_strength}, Max={max_strength}")
```
